```python
# import Delphos markers

import pandas as pd


markers_delpfos_df = pd.read_table("/Users/romain/Desktop/test_delphos/events.tsv",
                                   sep="\t", header=0)

markers_df = pd.DataFrame({
    "channel": markers_delpfos_df['channel'],
    "sample": markers_delpfos_df['sample_index']})

del markers_delpfos_df
```

```python
# import TRC datas
import sys
sys.path.insert(0, '/Users/romain/Study/Rheins/Thalamus_Git/Thalamus_Rheins_2025/Utils')

from FromTRC_TO_EDF_NEO_Local_Dates_segments import convert_micromed_mneObject
from helper import make_bipolar

trc_file = '/Users/romain/Desktop/test_delphos/EEG_792_.TRC'
raw_mne = convert_micromed_mneObject(trc_file)
raw_mne = make_bipolar(raw_mne, sens = '2-1')
   
channel_names = raw_mne.ch_names
signals = raw_mne.get_data()
times = raw_mne.times

del trc_file, raw_mne
```

```python
import sys
from PySide6 import QtWidgets
from eeg_spike_editor_qt import EEGEditor
from IPython import get_ipython

get_ipython().run_line_magic("gui", "qt")

  

# --- Lancement de l'application Qt
app = QtWidgets.QApplication.instance()
if app is None:
   app = QtWidgets.QApplication(sys.argv)
    
editor = EEGEditor(
    signals=signals,
    times=times,
    channel_names=channel_names,
    markers_df = markers_df,
    window_sec=20,
    n_display=40
    )

editor.show()
editor.resize(1500, 800)
editor.move(50, 200)  
app.exec()
QtWidgets.QApplication.quit()
```
